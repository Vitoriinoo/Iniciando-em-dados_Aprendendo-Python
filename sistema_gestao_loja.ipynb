## Parte 1: Cadastro e Validação de Usuário

Primeiro, vamos criar e validar os dados do usuário.


```python
# Parte 1: Cadastro de Usuário
# Crie as seguintes variáveis para o usuário:
# - nome = "João Silva"
# - idade = 25
# - e_estudante = True
# - tem_carteira_biblioteca = True
# - tem_multa_pendente = False
# Imprima os dados formatados
# Escreva seu código aqui

nome = "João Silva"
idade = 25
e_estudante = True
tem_carteira_biblioteca = True
tem_multa_pendente = False

texto = f"""
{"="*50}

Nome: {nome}
Idade: {idade}
É estudante? {e_estudante}
Tem carteira da Biblioteca? {tem_carteira_biblioteca}
Tem multa? {tem_multa_pendente}

{"="*50}
"""

print(texto)
```

## Parte 2: Sistema de Empréstimo

Agora vamos simular um empréstimo de livro.


```python
# Parte 3: Sistema de Empréstimo
# Crie variáveis para o empréstimo:
# - nome_livro = "Python para Iniciantes"
# - dias_emprestimo = 12
# - prazo_maximo = 10 (regular)
# - prazo_estudante = 15
# Determine o prazo baseado no tipo de usuário
# Calcule se está em atraso ou quantos dias restam
# Escreva seu código aqui

nome_livro = "Python para Iniciantes"
dias_emprestimo = 12
prazo_maximo = 10
prazo_estudante = 15


if e_estudante:
  prazo_permitido = prazo_estudante
  tipo_usuario = "Estudante"

else:
  prazo_permitido = prazo_maximo
  tipo_usuario = "Regular"

# Verificar se há dias em atraso
dias_atraso = dias_emprestimo - prazo_permitido

print(dias_atraso)

if dias_atraso > 0:
  print(f"ATRASO: {dias_atraso}")
else:
  dias_restantes = prazo_permitido - dias_emprestimo
  print(f"Você possui {dias_restantes} dias restante.")
```

## Parte 3: Cálculo de Multa

Vamos calcular a multa baseada no atraso.


```python
# Parte 4: Cálculo de Multa
# Calcule a multa baseada no atraso:
# - multa_por_dia = R$ 2.00
# - Se atraso > 7 dias: multa aumentada em 50%
# - Se é estudante: desconto de 10%
# - Se não está em atraso: sem multa
# Escreva seu código aqui



if dias_atraso > 0:
  multa_por_dia = 2

  multa_base = dias_atraso * multa_por_dia
  if dias_atraso > 7:
    multa_base = multa_base * 1.5
    print("Você possui mais de 7 dias de atraso. A sua multa aumenta em 50%")
  else:
    print(f"Você vai pagar o valor de multa padrão")

  desconto = 0

  if e_estudante:
    desconto = multa_base * 0.10
    print(f"Você por ser estudante tem 10% de desconto na multa")

    multa_final = multa_base - desconto

    print(f"A sua multa é de R${multa_final}. ")

else:
  multa_final = 0
  print(f"A sua multa final é de R${multa_final},00")
```

## Parte 4: Sistema de Renovação

Vamos verificar se o livro pode ser renovado.


```python
# Parte 5: Sistema de Renovação
# Crie variáveis: numero_renovacoes = 1, tem_reserva = False
# Verifique se pode renovar considerando:
# - Não pode estar em atraso
# - Máximo de 2 renovações
# - Não pode ter reserva
# - Não pode ter multa pendente
# Se pode renovar, calcule o novo prazo
# Escreva seu código aqui

# Sistema de Renovação
numero_renovacoes = 1 # Já renovou 1 vez
tem_reserva = False # Alguém reservou o livro

# verificar condições para renovação:
pode_renovar = True
motivo_impedimento = ""

# Verificar se esta em atraso:
if dias_atraso > 0:
  pode_renovar = False
  motivo_impedimento = "Livro em atraso - Devolva primeiro"
elif numero_renovacoes  >= 2:
  pode_renovar = False
  motivo_impedimento = "Limite de renovações atingido (máximo: 2)"
elif tem_reserva:
  pode_renovar = False
  motivo_impedimento = "Livro reservado por outro usuário"
elif tem_multa_pendente:
  pode_renovar = False
  motivo_impedimento = "Multa pendente - Deve ser paga primeiro"

# Exibir resultado
if pode_renovar:
  print("-Renovação PERMITIDA")





```

## Parte 4: Relatório Final

Vamos gerar um relatório resumido.


```python
# Parte 6: Relatório Final
# Crie um relatório completo exibindo:
# - Dados do usuário
# - Status de elegibilidade
# - Informações do empréstimo
# - Valor da multa (se houver)
# - Status de renovação
# Use formatação para deixar o relatório organizado
# Escreva seu código aqui

# Gerar Relatório Final
print("\n" + "=" * 60)
print("RELATÓRIO COMPLETO DA BIBLIOTECA".center(60))
print("=" * 60)

# Dados Usuário
print("\n - DADOS DO USUÁRIO: ")
print(f"Nome: {nome} | Idade: {idade} anos")
if e_estudante:
  print("     TIPO: Estudante")

else:
  print("     TIPO: Usuário Regular")

# STATUS
print("\n - STATUS: ")
if eligibilidade:
  print("   Eligibilidade: APROVADO")
else:
  print(f"   Eligibilidade: NEGADO - {motivo_impedimento}")

# EMPRÉSTIMO
print("\n - EMPRÉSTIMO: ")
print(f"   Livro: {nome_livro}")
print(f"   Dias desde empréstimo: {dias_emprestimo} | Prazo: {prazo_permitido} dias. ")

if dias_atraso > 0:
  print(f"     Status: ATRASADO ({dias_atraso} dias)")
else:
  dias_restantes = prazo_permitido - dias_emprestimo
  print(f"     Status: NO PRAZO ({dias_restantes} dias) restantes. ")

# MULTA
print("\n - MULTA:")
if multa_final > 0:
  print(f"    valor: R${multa_final:..2f}")
else:
  print(f"    valor: R$0,00 - Sem Multa")

# RENOVAÇÃO
print("\n - RENOVAÇÃO: ")
if pode_renovar:
  print(f"    Status: DISPONÍVEL ({numero_renovacoes}/2 renovações)")
else:
  print(f"    Status: INDISPONÍVEL ({motivo_impedimento}")

print("\n" + "="*60)
```

## Desafio: Seu Turno!

Agora é sua vez de criar um sistema! Siga os passos abaixo:

### Tarefa:
Crie um sistema de gestão de loja que:
1. Valida dados do cliente (idade >= 18, tem cadastro)
2. Calcula desconto baseado em:
   - Valor da compra >= 100: 5%
   - Valor da compra >= 200: 10%
   - Valor da compra >= 500: 15%
   - Se for cliente VIP: +5% adicional
   - Desconto máximo: 25%
3. Calcula frete:
   - Distância < 50 km: R$ 10
   - Distância 50-200 km: R$ 20
   - Distância > 200 km: R$ 30
   - Se compra >= 200: frete grátis
4. Gera relatório resumido

### Dicas:
- Use apenas estruturas condicionais (`if`, `elif`, `else`)
- Use condições aninhadas quando necessário
- Nomes de variáveis descritivos


```python
# ============================================
# DESAFIO: Sistema de Gestão de Loja
# ============================================

# 1. Crie dados do cliente:
#    - nome, idade, e_cliente_vip, tem_cadastro
nome = "Vitor Vitorino"
idade = 19
e_cliente_vip = True
tem_cadastro = True

# 2. Crie dados da compra:
#    - valor_compra, distancia
valor_compra = 500
distancia = 255

# 3. Valide o cliente (idade >= 18, tem cadastro)

if idade >= 18 and tem_cadastro:
  print("CADASTRO OK")

# 4. Calcule desconto baseado nas regras especificadas
if valor_compra >= 500:
  desconto = valor_compra * 0.15
elif valor_compra >= 200:
  desconto = valor_compra * 0.10
elif valor_compra >= 100:
  desconto = valor_compra * 0.05
else:
  desconto = 0

desconto_vip = valor_compra * 0.05

if e_cliente_vip:
  desconto = desconto + desconto_vip

if desconto > (valor_compra * 0.25):
    desconto = valor_compra * 0.25

# 5. Calcule frete baseado na distância
if valor_compra >= 200:
    frete = 0
else:
    if distancia < 50:
        frete = 10
    elif 50 <= distancia <= 200:
        frete = 20
    else:
        frete = 30

# 6. Calcule o valor final (valor - desconto + frete)
valor_da_compra = valor_compra - desconto + frete

# 7. Gere um relatório completo e organizado

# Escreva seu código aqui

print("="*60)
print("RELATÓRIO COMPLETO".center(60))
print("="*60)
print("- DADOS DO CLIENTE")
print(f"""Nome: {nome}
Idade: {idade}
Cliente vip: {e_cliente_vip}
Tem Cadastro: {tem_cadastro}""")
print("-"*60)
print("- DADOS DA COMPRA")
print(f"Valor da compra: R${valor_compra:.2f}")
print(f"Distância: {distancia} km")
print("-"*60)
print("- DESCONTO DISPONÍVEL")
print(f"Valor do desconto disponível: R${desconto:.2f} \n(clientes vips recebem 5% a mais de desconto além do padrão)")
print("-"*60)
print("- VALOR FINAL")
print(f"O valor final da compra ficou em: R${valor_da_compra:.2f}")
print("="*60)
```
