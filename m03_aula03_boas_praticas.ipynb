## 1. Por que Código Legível é Importante?

Código legível é:
- **Fácil de entender** - você e outros programadores conseguem ler
- **Fácil de manter** - é simples fazer alterações
- **Menos propenso a erros** - menos confusão = menos bugs
- **Profissional** - demonstra boas práticas de programação

**Lembre-se:** Código é lido muito mais vezes do que escrito!


## 2. Boas Práticas para Estruturas Condicionais

### 2.1 Use Nomes de Variáveis Descritivos

**Ruim:**
```python
if x > 18:
    print("OK")
```

**Bom:**
```python
if idade >= 18:
    print("Você é maior de idade")
```


````python
# Exemplo: Comparação de código
# Escreva DUAS versões do mesmo código:
# RUIM:
x = 22
if x > 20:
  print("Maior")
# VERSÃO 1 (RUIM): Use variáveis com nomes ruins (n, s, etc.)
# VERSÃO 2 (BOM): Use nomes descritivos (nota, status, etc.)
# BOM:
idade = 22
if idade > 20:
  print("Maior")
# Compare a legibilidade dos dois códigos
# Escreva seu código aqui
````

### 2.2 Evite Condições Negativas Quando Possível

**Ruim:**
```python
if not idade < 18:
    print("Maior de idade")
```

**Bom:**
```python
if idade >= 18:
    print("Maior de idade")
```


````python
# Exemplo: Evitar condições negativas
# Crie uma variável idade = 20
# Escreva DUAS versões comparando:
# VERSÃO 1: Use 'not' na condição (if not idade < 18)
# VERSÃO 2: Use condição positiva (if idade >= 18)
# Observe qual é mais fácil de ler
# Escreva seu código aqui

# versão 1 not:
idade = 22
if not idade < 18:
  print("Maior de idade")

# versão 2
if idade > 18:
  print("Maior de idade")
````

### 2.3 Use Early Return (Retorno Antecipado)

Quando possível, trate casos especiais primeiro e retorne cedo. Isso reduz aninhamento.

**Ruim:**
```python
if condicao:
    if outra_condicao:
        # código principal
```

**Bom:**
```python
if not condicao:
    return
if not outra_condicao:
    return
# código principal
```


````python
# Exemplo: Reduzir aninhamento (early return)
# Crie variáveis: idade = 20, tem_carteira = True
# Escreva DUAS versões:
# VERSÃO 1: Use condições aninhadas (if dentro de if)
# VERSÃO 2: Use early return (trate casos especiais primeiro)
# Compare qual é mais clara
# Escreva seu código aqui

idade = 20
tem_carteira = True

if idade >= 18: #essa é a condiçção principal
  if tem_carteira:
    print("Você pode dirigir")
  else:
    print("Você não pode dirigir")
else:
    print("Você precisa ter 18 anos")


# Forma mais eficiente:

if idade < 18:
  print("Você precisa ter 18 anos")

elif tem_carteira:
  print("Você pode dirigir")

else:
    print("Você precisa de carteira")
````

### 2.4 Use Operadores Lógicos para Simplificar

**Ruim:**
```python
if idade >= 18:
    if tem_carteira:
        print("Pode dirigir")
```

**Bom:**
```python
if idade >= 18 and tem_carteira:
    print("Pode dirigir")
```


````python
# Exemplo: Simplificar com operadores lógicos
# Crie variáveis: idade = 20, tem_carteira = True
# Escreva DUAS versões:
# VERSÃO 1: Use if aninhado
# VERSÃO 2: Use operador 'and' em uma única condição
# Compare qual é mais simples
# Escreva seu código aqui
idade = 20
tem_carteira = True

if idade >= 18:
   if tem_carteira:
    print("Você pode dirigir")

if idade >= 18 and tem_carteira:
  print("Você pode dirigir")

else:
  print("Você não pode dirigir")
````

- Use um editor que mostre a indentação


````python
# Exemplo: Praticar indentação correta
# Crie uma variável idade = 25
# Crie uma estrutura if-else com condição aninhada
# Use indentação de 4 espaços para cada nível
# Observe a diferença entre os níveis de indentação
# Escreva seu código aqui

idade = 66

if idade >=18:
  print("Você é maior de idade.")
  if idade >= 60:
      print("Você é maior de 60")
````

**Bom:**



````python
# Exemplo: Escrever comentários úteis
# Crie variáveis: nota = 8.5, frequencia = 85
# Escreva um código que verifica aprovação
# Adicione comentários que explicam o "POR QUÊ" (não o "o quê")
# Exemplo: explique a regra de negócio, não o que o código faz
# Escreva seu código aqui
nota = 8.5
frequencia = 85

# aqui precisamos verificar se o aluno possui nota superior a 7
# e se sua frequencia é maior do que 70.
if nota >= 7 and frequencia >= 70:
  print("Você foi aprovado")

````

## 3. Erros Comuns e Como Evitá-los

### 3.1 Esquecer os Dois Pontos (:)

**Erro:**
```python
if idade >= 18  # Falta o :
    print("Maior")
```

**Correto:**
```python
if idade >= 18:  # Tem o :
    print("Maior")
```


### 3.2 Usar = em vez de ==

**Erro:**
```python
if idade = 18:  # = é atribuição, não comparação!
    print("Tem 18 anos")
```

**Correto:**
```python
if idade == 18:  # == é comparação
    print("Tem 18 anos")
```


### 3.3 Indentação Incorreta

**Erro:**
```python
if idade >= 18:
print("Maior")  # Falta indentação!
```

**Correto:**
```python
if idade >= 18:
    print("Maior")  # Com indentação correta
```
